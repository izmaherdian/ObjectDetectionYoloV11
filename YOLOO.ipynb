```python
from pycocotools.coco import COCO
import requests
import os
import tqdm

# Path ke file anotasi COCO
ann_file = 'annotations_trainval2017/annotations/instances_val2017.json'  # Ganti dengan path yang sesuai
desired_categories = ['person', 'car', 'chair', 'table']  # Daftar kategori yang diinginkan
filtered_image_dir = 'FilterImage'  # Direktori untuk menyimpan gambar terfilter
os.makedirs(filtered_image_dir, exist_ok=True)

# Inisialisasi API COCO
coco = COCO(ann_file)

# Ambil ID kategori berdasarkan nama kategori yang diinginkan
cat_ids = coco.getCatIds(catNms=desired_categories)

# Ambil ID gambar yang mengandung kategori tersebut
img_ids = coco.getImgIds(catIds=cat_ids)
images = coco.loadImgs(img_ids)

# Fungsi untuk menampilkan persentase dan sisa MB
def download_image_with_progress(img_url, file_path):
    img_data = requests.get(img_url, stream=True)  # Menggunakan stream agar dapat mendownload secara bertahap
    total_size = int(img_data.headers.get('content-length', 0)) / 1024 / 1024  # Ukuran dalam MB
    downloaded = 0

    with open(file_path, 'wb') as handler:
        for data in img_data.iter_content(chunk_size=1024):
            downloaded += len(data)
            handler.write(data)
            percent = (downloaded / total_size) * 100
            remaining_mb = total_size - (downloaded / 1024 / 1024)
            print(f"\rDownloading: {percent:.2f}% - {remaining_mb:.2f} MB remaining", end='')

# Download gambar sesuai kategori dengan persentase progress
for img in tqdm.tqdm(images, desc="Downloading images", unit="image"):
    img_url = img['coco_url']  # URL gambar dari COCO
    file_path = os.path.join(filtered_image_dir, img['file_name'])
    
    # Download gambar dan tampilkan persentase
    download_image_with_progress(img_url, file_path)
    print()  # Line break setelah setiap gambar selesai diunduh
```
