```python
import cv2
from ultralytics import YOLO

# Load model yang sudah dilatih (best.pt)
model = YOLO('best.pt')  # Pastikan best.pt adalah path yang benar untuk model Anda

# Inisialisasi kamera eksternal (biasanya ID kamera eksternal adalah 1, 2, dll.)
cap = cv2.VideoCapture(0)  # Ganti angka 1 sesuai ID kamera eksternal Anda

# Cek apakah kamera berhasil dibuka
if not cap.isOpened():
    print("Tidak dapat mengakses kamera eksternal.")
    exit()

# Loop untuk mendapatkan frame dari kamera dan melakukan prediksi
while True:
    ret, frame = cap.read()
    if not ret:
        print("Gagal mendapatkan frame dari kamera.")
        break

    # Lakukan prediksi pada frame
    results = model(frame)

    # Menggambar bounding boxes dan label pada frame
    annotated_frame = results[0].plot()

    # Tampilkan frame dengan hasil prediksi
    cv2.imshow("YOLO Prediction", annotated_frame)

    # Keluar dengan menekan 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Lepaskan sumber daya
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Load the model
model = YOLO('best.pt')  # Update path to where you saved the model locally

# Run real-time detection
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    annotated_frame = results[0].plot()

    cv2.imshow('YOLO Real-Time Detection', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
